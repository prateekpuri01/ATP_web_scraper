```python
import csv
    
def check_p1_serve(row):
    fsp=int(row[8])
    fsm=int(row[9])
    fsh=int(row[10])
    if (fsp-fsm/fsh)>.03:
        return True
    else:
        return False
    
def check_p2_serve(row):
    fsp=int(row[17])
    fsm=int(row[18])
    fsh=int(row[19])
    if (fsp-fsm/fsh)>.03:
        return True
    else:
        return False
    
def check_basics(row):
    if (row[10]=='0' or row[11]=='0' or row[12]=='0' 
        or row[14]=='0' or row[18]=='0' or row[19]=='0' or row[20]=='0' 
        or row[21]=='0'):
        
        return True
    else:
        return False
    

        
def clean_data(year):
    print('Year: '+str(year))
    file_name='match_data_'+str(year)+'.csv'
    cleaned_lines=[]
    with open(file_name,'r') as csvfile:
        readCSV = csv.reader(csvfile, delimiter=',')
        loop_track=0
    
        for i,row in enumerate(readCSV):
            loop_track+=1
            if i==0:
                continue
            if check_basics(row):
                continue          
            elif check_p1_serve(row):
                temp_row=row
                temp_row[9]=int(temp_row[8])*int(temp_row[10])
                cleaned_lines.append(temp_row)
            elif check_p2_serve(row):
                temp_row=row
                temp_row[18]=int(temp_row[17])*int(temp_row[19])
                cleaned_lines.append(temp_row)
            else:
                cleaned_lines.append(row)
                
    print("Num initial rows: {}".format(loop_track))
    print("Num rows after cleaning: {}".format(len(cleaned_lines)))
    cleaned_file_name='match_data_cleaned_'+str(year)+'.csv'
    
    with open(cleaned_file_name,'a',newline='') as csvFile:
            for line in cleaned_lines:
                writer = csv.writer(csvFile)
                writer.writerow(line)
    return None
```

```python
for year in range(1991,2020):
    clean_data(year)
```
